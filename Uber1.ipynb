# Rename Columns

```python
print("BEFORE — Column names:")
print(df.columns.tolist())

# Pipeline: lowercase → strip whitespace → replace spaces with underscores
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(' ', '_', regex=False)
)

print("\nAFTER — Standardized column names:")
print(df.columns.tolist())
```

# Check Duplication

```python
total_dups = df.duplicated().sum()
print(f"Total fully duplicate rows: {total_dups}")

# If duplicates exist, inspect them before dropping
if total_dups > 0:
    print("Sample of duplicate rows:")
    print(df[df.duplicated(keep=False)])
```

# Check Missing Values

```python
missing = df.isna().sum()
missing_pct = (df.isna().sum() / len(df) * 100).round(2)

missing_report = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
}).sort_values('Missing %', ascending=False)


print("Missing Value Report:")
print(missing_report[missing_report['Missing Count'] > 0].to_string())
```

```python
for col in ['reason_for_cancelling_by_customer', 'driver_cancellation_reason', 'incomplete_rides_reason']:
    df[col] = df[col].fillna('Not Applicable')
    print(f"✅ '{col}' filled with 'Not Applicable'")
```

```python
for col in ['cancelled_rides_by_customer', 'cancelled_rides_by_driver', 'incomplete_rides', 'booking_value', 'ride_distance']:
    df[col] = df[col].fillna('0')
    print(f"✅ '{col}' filled with '0'")
```

```python
for col in ['payment_method']:
    df[col] = df[col].fillna('No Payment')
    print(f"✅ '{col}' filled with 'No Payment'")
```
