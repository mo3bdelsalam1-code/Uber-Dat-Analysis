## Data Visualization

```python

df["Booking Status"].value_counts().plot(kind="bar")

plt.title("Booking Status Distribution")
plt.xlabel("Booking Status")
plt.ylabel("Number of Bookings")

plt.show()
```

```python

df["Vehicle Type"].value_counts().plot(kind="bar")

plt.title("Vehicle Type Distribution")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Bookings")

plt.show()
```

```python
df["Payment Method"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Payment Method Distribution")
plt.ylabel("")

plt.show()
```

```python

df["Booking Value"].plot(kind="hist", bins=20)

plt.title("Booking Value Distribution")
plt.xlabel("Booking Value")
plt.ylabel("Frequency")

plt.show()
```

```python
df["Ride Distance"].plot(kind="hist", bins=20)

plt.title("Ride Distance Distribution")
plt.xlabel("Ride Distance")
plt.ylabel("Frequency")

plt.show()
```
